```python
from skimage.io import imread, imsave
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt

for l in os.listdir():
    os.chdir(l)
    for k in os.listdir():
        os.chdir(k)
        for image in os.listdir():
            # reading the image
            img = imread(image)
            #plt.axis("off")
            #plt.imshow(img)
            #print(img.shape)

            #creating hog features

            fd, hog_image = hog(img, orientations=9, pixels_per_cell=(9, 9),cells_per_block=(1, 1), visualize=True, multichannel=False)


            imsave(image, hog_image)
        os.chdir('..')
    os.chdir('..')
```
